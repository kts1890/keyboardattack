```python
import os
import glob
import librosa
import numpy as np
import pandas as pd
import pickle

import tensorflow as tf
from tensorflow import keras
from keras import layers
from keras.utils import to_categorical

import sklearn.preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
def extract_feature(file_name,text):
  file = open(text, 'r')
  a=file.readlines()
  b=[]
  data_dic=['q','w','e','r','t','y','u','i','o','p','a','s','d','f','g','h','j','k','l','z','x','c','v','b','n','m']
  df=pd.DataFrame(data_dic)
  for i in a:
    b.append(i.replace('\n',''))
  print('file name :', file_name)
  trainset=[]
  audio, sample_rate = librosa.load(file_name,sr=44100)
  print(len(audio))
  for i in range(int(librosa.get_duration(y=audio, sr=sample_rate)/2)):
    print(i)
    audios=audio[i*sample_rate*2:i*sample_rate*2+sample_rate*2]
    print(len(audios))
    mfccs = librosa.feature.mfcc(y=audios, sr=sample_rate, n_mfcc=40)
    mfccs = sklearn.preprocessing.scale(mfccs, axis=1)
    print(mfccs.shape)
    trainset.append([mfccs,data_dic.index(b[i])])

  return trainset
```

```python
trainset=[]
trainset+= extract_feature('/content/a.wav','/content/a.txt')
trainset+= extract_feature('/content/b.wav','/content/b.txt')
trainset+= extract_feature('/content/c.wav','/content/c.txt')
trainset+= extract_feature('/content/d.wav','/content/d.txt')
trainset+= extract_feature('/content/e.wav','/content/e.txt')
trainset+= extract_feature('/content/f.wav','/content/f.txt')
trainset+= extract_feature('/content/g.wav','/content/g.txt')
trainset+= extract_feature('/content/h.wav','/content/h.txt')
trainset+= extract_feature('/content/i.wav','/content/i.txt')
trainset+= extract_feature('/content/j.wav','/content/j.txt')
trainset+= extract_feature('/content/k.wav','/content/k.txt')
trainset+= extract_feature('/content/l.wav','/content/l.txt')
trainset+= extract_feature('/content/m.wav','/content/m.txt')
trainset+= extract_feature('/content/n.wav','/content/n.txt')
trainset+= extract_feature('/content/o.wav','/content/o.txt')
trainset+= extract_feature('/content/p.wav','/content/p.txt')
trainset+= extract_feature('/content/q.wav','/content/q.txt')
trainset+= extract_feature('/content/r.wav','/content/r.txt')
trainset+= extract_feature('/content/s.wav','/content/s.txt')
trainset+= extract_feature('/content/t.wav','/content/t.txt')
trainset+= extract_feature('/content/u.wav','/content/u.txt')
trainset+= extract_feature('/content/v.wav','/content/v.txt')
trainset+= extract_feature('/content/w.wav','/content/w.txt')
trainset+= extract_feature('/content/x.wav','/content/x.txt')
trainset+= extract_feature('/content/y.wav','/content/y.txt')
trainset+= extract_feature('/content/z.wav','/content/z.txt')
```

```python
librosa.display.specshow(trainset[6][0],sr=44100,x_axis='time')
```

```python
#data_dic = {'label' : ['q','w','e','r','t','y','u','i','o','p','a','s','d','f','g','h','j','k','l','z','x','c','v','b','n','m']}

train_mfccs = [a for (a,b) in trainset]
train_y = [b for (a,b) in trainset]

train_mfccs = np.array(train_mfccs)
y_train = to_categorical(np.array(train_y))

x_train, x_valid, y_train, y_valid = train_test_split(train_mfccs, y_train, test_size=0.2, shuffle=True, stratify=y_train, random_state=34)
print(x_train.shape)
print(y_train.shape)
print(x_valid.shape)
x_train_subset = x_train[:200]
y_train_subset = y_train[:200]
# x_train_subset과 x_valid를 합치기 위해 축을 지정하여 concatenate 함수 사용
x_valid = np.concatenate((x_train_subset, x_valid), axis=0)
y_valid = np.concatenate((y_train_subset, y_valid), axis=0)

x_train = np.expand_dims(x_train, -1)
x_valid = np.expand_dims(x_valid, -1)
print(x_valid.shape)
```

```python
n_columns = 173
n_row = 40
n_channels = 1
n_classes = 26

model = keras.Sequential()

model.add(layers.Conv2D(input_shape=(n_row, n_columns, n_channels), filters=16, kernel_size=2, activation='relu'))
model.add(layers.MaxPooling2D(pool_size=2))
model.add(layers.Conv2D(kernel_size=2, filters=32, activation='relu'))
model.add(layers.MaxPooling2D(pool_size=2))

model.add(layers.Conv2D(kernel_size=2, filters=64, activation='relu'))
model.add(layers.MaxPooling2D(pool_size=2))
model.add(layers.Conv2D(kernel_size=2, filters=128, activation='relu'))
model.add(layers.MaxPooling2D(pool_size=2))

model.add(layers.Flatten())
model.add(tf.keras.layers.Dense(units=26, activation='softmax'))

model.summary()
```

```python
training_epochs = 30
num_batch_size = 128

learning_rate = 0.001
opt = keras.optimizers.Adam(learning_rate=learning_rate)

model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
history = model.fit(x_train, y_train, validation_data = (x_valid, y_valid), batch_size=num_batch_size, epochs=training_epochs)
#history = model.fit(x_train, y_train, batch_size=num_batch_size, epochs=training_epochs)
```

```python
results = model.evaluate(x_train, y_train, batch_size=128)
print('test loss, test acc:', results)
```

```python
# 훈련 및 검증 손실 그래프
import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# 훈련 및 검증 정확도 그래프

plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
predictions = model.predict(x_valid)
predicted_labels = np.argmax(predictions, axis=1)
real_labels=np.argmax(y_valid,axis=1)
all=np.zeros(26)
result = np.zeros(26)
for i in range(452):
  all[real_labels[i]]+=1
  if(predicted_labels[i]==real_labels[i]):
    result[real_labels[i]]+=1
print(result/all)
print(all)
```
